```python
#Importation des bibliothèques

import pandas as pd
from scipy.stats import f_oneway, chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Lecture du dataset
data = pd.read_csv('german_credit_data.csv')
data
```

```python
# Exploration initiale
print(data.info())
print(data.describe())
```

```python
# Matrice de corrélation
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Matrice de Corrélation")
plt.show()
```

```python
data['Risk'].unique()
# categorical_columns = [col for col in data.drop('Risk', axis = 1).columns if data[col].dtype == 'object']
# categorical_columns
```

```python
categorical_columns = [col for col in data.drop('Risk', axis = 1).columns if data[col].dtype == 'object']

# Créer une table de contingence
for value in categorical_columns:
    contingency_table = pd.crosstab(data[value], data['Risk'])
    print("Table de contingence:\n", contingency_table)

    # Test Chi-square
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)

    print("\nStatistique Chi²:", chi2)
    print("P-value:", p_value)
    print("Degrés de liberté:", dof)
    print("Fréquences attendues:\n", expected)

    # Interprétation
    if p_value < 0.05:
        print("Il existe une association significative entre ", value," et le statut de crédit.")
    else:
        print("Aucune association significative entre ", value," et le statut de crédit.")
```

```python
#Test chi2 sur les variables catégorielles

categorical_columns = [col for col in data.drop('Risk', axis = 1).columns if data[col].dtype == 'object']

# Séparer les groupes
for value in categorical_columns:
    group1 = data[data['Risk'] == 'good'][value]
    group2 = data[data['Risk'] == 'bad'][value]

    # Test ANOVA
    stat, p_value = f_oneway(group1, group2)

    print("Statistique F:", stat)
    print("P-value:", p_value)

    # Interprétation
    if p_value < 0.05:
        print("Les moyennes de revenu diffèrent significativement entre les groupes.")
    else:
        print("Aucune différence significative entre les moyennes des groupes.")
```
